```python
from tokenizers import Tokenizer,models,trainers,pre_tokenizers
data=['hello my name is musheer','what are you doing','chalti kya khandala']

tokenizer=Tokenizer(models.BPE())
tokenizer.pre_tokenizers = pre_tokenizers.Whitespace()
trainer=trainers.BpeTrainer(vocab_size=100)
tk=tokenizer.train_from_iterator(data,trainer)
tokenizer.save('my-tokenizer.json')
```

```python
import torch
from transformers import AutoModel , AutoTokenizer
tokenizer=AutoTokenizer.from_pretrained('bert-base-cased')
bert_model=AutoModel.from_pretrained('bert-base-uncased')
bert_model
```

```python
input=tokenizer('hello world ,',return_tensors='pt')
with torch.no_grad():
  output=bert_model(**input)
print(output)
```

```python
from torch.nn.functional import cosine_similarity
sen1=['hello my name is musheer']
sen2=['hello my self musheer']

input1=tokenizer(sen1,return_tensors='pt')
input2=tokenizer(sen2,return_tensors='pt')

with torch.no_grad():
  output1=bert_model(**input1).last_hidden_state[0].mean(dim=0)
  output2=bert_model(**input2).last_hidden_state[0].mean(dim=0)

similarity=cosine_similarity(output1,output2,dim=0)
print('Similarity : ',round(similarity.item()*100,2),'%')
```

```python
from sentence_transformers import SentenceTransformer
model=SentenceTransformer('all-MiniLM-L6-v2')
sen1='hello my name is musheer'
sen2='hello my self musheer'
emb1=model.encode(sen1,convert_to_tensor=True)
emb2=model.encode(sen2,convert_to_tensor=True)

similarity = cosine_similarity(emb1, emb2, dim=0)
print(f"Similarity: {similarity.item()*100:.2f} %")
```
